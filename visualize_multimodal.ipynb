```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

# Load merged multimodal dataset
df = pd.read_csv("LUAD_multimodal_dataset_with_paths.csv")

# Pick one patient to visualize
example = df.iloc[0]  # change index to view others
nii_path = example["ImagePath"]
report_text = example["Report"]

print(f"üìÑ Patient ID: {example['PatientID']}")
print(f"üß† Survival Label: {example['survival_label']}")
print(f"üìÇ Image Path: {nii_path}")
print(f"\nü©ª Radiology Report:\n{report_text}")

# Load and visualize middle CT slice
if os.path.exists(nii_path):
    img = nib.load(nii_path)
    data = img.get_fdata()
    mid_slice = data[:, :, data.shape[2] // 2]

    plt.figure(figsize=(8, 6))
    plt.imshow(np.rot90(mid_slice), cmap="gray")
    plt.title(f"{example['PatientID']} | {example['survival_label']}")
    plt.axis("off")
    plt.show()
else:
    print("‚ö†Ô∏è Image file not found.")
```
